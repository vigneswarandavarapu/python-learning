```python
marks=[12,34,56,78,89,678]
print(f'Total = {sum(marks)}')
print(f'Max = {max(marks)}')
print(f'Min = {min(marks)}')
print(f'Length = {len(marks)}')
print(f'Average = {sum(marks)/len(marks)}')
```

```python
marks=[12,34,56,78,89,678]
sum=0;
length=0;
max=marks[0];
min=marks[0];

for i in marks:
  sum=sum+i;
  length=length+1;
  if i > max:
    max=i
  if i < min:
    min=i

print(f'Total = {sum}')
print(f'Max = {max}')
print(f'Min = {min}')
print(f'Length = {length}')
print(f'Average = {sum/length}')
```
