```python
def addition(a,b=10):
  print(f'values of a : {a}')
  print(f'values of b : {b}')
  return a+b;

print(addition(10))
```

```python
def addition(a,b=10,c=20):
  print(f'values of a : {a}')
  print(f'values of b : {b}')
  print(f'values of c : {c}')
  return a+b+c;

print(addition(10,c=10))
```
