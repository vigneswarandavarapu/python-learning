````python
#Find the largest number in a list.

list =[10,100,12,43,56,78,65]
max=list[0]
for i in list:
  if max<i:
    max=i;
else:
  print(f'Max number : {max}');
````

````python
#Count even and odd numbers.
list =[10,100,12,43,56,78,65]
evenNumbers=[];
oddNumbers=[];

for i in list:
  if i%2==0:
    evenNumbers.append(i);
  else:
    oddNumbers.append(i);

print(f'Even Numbers : {evenNumbers} And count is : {len(evenNumbers)}');
print(f'Odd Numbers : {oddNumbers} And count is :{len(oddNumbers)}');
````

````python
#Reverse a list

list =[10,100,12,43,56,78,65]

reversedList=[];

for i in range(1,len(list)+1):
  reversedList.append(list[-i]);

print(f'Reversed List : {reversedList}')
````

````python
#Find the sum and average of marks.
list =[10,100,12,43,56,78,65]

sum=0;

for i in list:
  sum=sum+i;
else:
  print(f'sum of numbers: {sum}');
  print(f'average of numbers: {sum/len(list)}');
````

````python
#Check whether a number is prime.

number=int(input('Enter a number : '));

for i in range(2,number-1):
  if number%i!=0:
    continue;
  else:
    print('Not Prime Number');
    break;
else:
  print('Prime Number')
````

````python
#Generate Fibonacci series.
number=int(input('Enter a number : '));

fabSeries=[];

for i in range(0,number):
  if len(fabSeries)<2:
    fabSeries.append(i);
    continue;

  fabSeries.append(fabSeries[i-1]+fabSeries[i-2]);

print(f'Fibonacci Series : {fabSeries}');
````
